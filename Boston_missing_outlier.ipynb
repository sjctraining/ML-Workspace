### Outliers

```python
var_out = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
Q1 = df.quantile(.25); Q3 = df.quantile(.75); IQR = Q3 - Q1
```

```python
LB = Q1 - IQR * 1.5
```

```python
UB = Q3 + IQR * 1.5
```

```python
IQR = Q3 - Q1
```

```python
select = (df[var_out] > LB[var_out]) & (df[var_out] < UB[var_out])
```

```python
type(select)
```

```python
select[['CHAS', 'ZN']] = True
```

```python
df[select]
```

```python
df.describe()
```

```python
#df = df.dropna()
#df = df.fillna(df.mean())
#df = df.fillna(df.median())
df = df[df > LB].fillna(LB); df = df[df < UB].fillna(UB)
```

```python
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
y = df['MEDV']
```

### Import the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

#### Fit or Train the model

```python
lm.fit(x_train, y_train)
```

```python
lm.score(x_train, y_train)
```

```python
lm.score(x_test, y_test)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
y_pred = lm.predict(x_test)
```

```python
plt.scatter(y_pred, y_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
'R^2 =',metrics.explained_variance_score(y_test,y_pred)
```

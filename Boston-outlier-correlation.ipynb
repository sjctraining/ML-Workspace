```python
Q1 = df[var].quantile(.25); Q3 = df[var].quantile(.75); IQR = Q3 - Q1
```

```python
LB = Q1 - IQR * 1.5
```

```python
LB
```

```python
UB = Q3 + IQR * 1.5
```

```python
UB
```

```python
select = (df[var] > LB[var]) & (df[var] < UB[var])
```

```python
type(select)
```

```python
select['chas'] = True
```

```python
df = df[select]
```

```python
y = df['medv']
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

### Fit / Train the model

```python
lm.fit(x_train, y_train)
```

```python
lm.score(x_train, y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
lm.score(x_test, y_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
'R^2 =',metrics.explained_variance_score(y_test,y_pred)
```
